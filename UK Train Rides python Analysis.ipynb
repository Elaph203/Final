```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df['Journey Duration (min)'], df['Delay (min)'], alpha=0.6)
plt.xlabel('Journey Duration (min)')
plt.ylabel('Delay (min)')
plt.title('Data with Negative Values')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
negative_data = df[df['Journey Duration (min)'] < 0]
plt.scatter(negative_data['Journey Duration (min)'], negative_data['Delay (min)'], alpha=0.8, color='red')
plt.xlabel('Journey Duration (min)')
plt.ylabel('Delay (min)')
plt.title('Negative Values Highlighted')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Negative values found: {len(negative_data)}")
print(f"Range: {negative_data['Journey Duration (min)'].min()} to {negative_data['Journey Duration (min)'].max()} minutes")
```

```python
# Detect and analyze errors
negative_count = len(df[df['Journey Duration (min)'] < 0])
negative_percentage = (negative_count / len(df)) * 100

print("Error Analysis:")
print(f"Negative values: {negative_count} ({negative_percentage:.2f}%)")
print(f"Min value: {df['Journey Duration (min)'].min()}")
print(f"Max value: {df['Journey Duration (min)'].max()}")
```

```python
# Fix the data
df_fixed = df.copy()
df_fixed['Journey Duration (min)'] = df_fixed['Journey Duration (min)'].abs()

print("Data Fix Results:")
print(f"Negative values before: {negative_count}")
print(f"Negative values after: {(df_fixed['Journey Duration (min)'] < 0).sum()}")
print(f"Data preserved: {len(df_fixed)} rows")
```

```python
# Clean visualization after fix
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df_fixed['Journey Duration (min)'], df_fixed['Delay (min)'], alpha=0.6, c=df_fixed['Delay (min)'], cmap='viridis')
plt.colorbar(label='Delay (min)')
plt.xlabel('Journey Duration (min)')
plt.ylabel('Delay (min)')
plt.title('Fixed Data - Clean Visualization')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
z = np.polyfit(df_fixed['Journey Duration (min)'], df_fixed['Delay (min)'], 1)
p = np.poly1d(z)
plt.scatter(df_fixed['Journey Duration (min)'], df_fixed['Delay (min)'], alpha=0.6)
plt.plot(df_fixed['Journey Duration (min)'], p(df_fixed['Journey Duration (min)']), "r-", linewidth=2)
plt.xlabel('Journey Duration (min)')
plt.ylabel('Delay (min)')
plt.title('With Trend Line')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

correlation = df_fixed['Journey Duration (min)'].corr(df_fixed['Delay (min)'])
print(f"Correlation coefficient: {correlation:.3f}")
```

```python
# Analyze relationship between purchase time and journey time
plt.figure(figsize=(12, 5))

# Convert time columns properly
df_fixed['Purchase_Hour'] = pd.to_datetime(df_fixed['Time of Purchase'], format='%H:%M:%S').dt.hour
df_fixed['Departure_Hour'] = pd.to_datetime(df_fixed['Departure Time'], format='%H:%M:%S').dt.hour

plt.subplot(1, 2, 1)
# How early people book before departure
df_fixed['Hours_Before_Departure'] = df_fixed['Departure_Hour'] - df_fixed['Purchase_Hour']
df_fixed['Hours_Before_Departure'] = df_fixed['Hours_Before_Departure'].apply(lambda x: x if x >= 0 else x + 24)
booking_times = df_fixed['Hours_Before_Departure'].value_counts().sort_index().head(24)
booking_times.plot(kind='bar', color='teal', alpha=0.7)
plt.title('Booking Time Before Departure')
plt.xlabel('Hours Before Departure')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
# Price vs booking time
price_by_booking_time = df_fixed.groupby('Hours_Before_Departure')['Price'].mean().head(24)
price_by_booking_time.plot(kind='line', color='coral', marker='o')
plt.title('Average Price vs Booking Time')
plt.xlabel('Hours Before Departure')
plt.ylabel('Average Price')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Analyze Railcard impact on pricing
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
# Average price by railcard type
price_by_railcard = df_fixed.groupby('Railcard')['Price'].mean().sort_values(ascending=False)
price_by_railcard.plot(kind='bar', color=['blue', 'green', 'red', 'orange'])
plt.title('Average Ticket Price by Railcard Type')
plt.xlabel('Railcard Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
# Price distribution by railcard
railcard_types = df_fixed['Railcard'].unique()
price_data = [df_fixed[df_fixed['Railcard'] == card]['Price'] for card in railcard_types]
plt.boxplot(price_data, labels=railcard_types)
plt.title('Price Distribution by Railcard Type')
plt.xlabel('Railcard Type')
plt.ylabel('Ticket Price')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Analyze delay patterns by day of week
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
# Average delay by day
delay_by_day = df_fixed.groupby('Day of Week')['Delay (min)'].mean()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
delay_by_day = delay_by_day.reindex(days_order)
delay_by_day.plot(kind='bar', color='lightcoral')
plt.title('Average Delay by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Delay (min)')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
# Percentage of delayed journeys by day
delayed_by_day = df_fixed.groupby('Day of Week')['Is Delayed'].apply(lambda x: (x == 'Yes').mean())
delayed_by_day = delayed_by_day.reindex(days_order)
delayed_by_day.plot(kind='bar', color='lightblue')
plt.title('Percentage of Delayed Journeys by Day')
plt.xlabel('Day of Week')
plt.ylabel('Percentage Delayed')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
